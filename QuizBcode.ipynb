```python
dSet1=dSet.drop(["date","lights"], axis=1)   #question 13
```

```python
dSet1.head()
```

```python
#Firstly, we normalise our dataset to a common scale using the min max scaler

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_dSet1 = pd.DataFrame(scaler.fit_transform(dSet1), columns=dSet1.columns)

#feature selection, selecting which column will be the predictor/independent variable and  the dependent/target variable
features_dSet1 = normalised_dSet1.drop(columns=[ 'T1', 'RH_1', 'T2', 'RH_2', 'T3',
       'RH_3', 'T4', 'RH_4', 'T5', 'RH_5', 'T6', 'RH_6', 'T7', 'RH_7', 'T8',
       'RH_8', 'T9', 'RH_9', 'T_out', 'Press_mm_hg', 'RH_out', 'Windspeed',
       'Visibility', 'Tdewpoint', 'rv1', 'rv2'])

heating_target = normalised_dSet1['Appliances']
```

```python
#Now, we split our dataset into the training and testing dataset. Recall that wehad earlier segmented the features and target variables.

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features_dSet1, heating_target,test_size=0.3, random_state=42)

#Training the model
from sklearn.linear_model import LinearRegression
linear_model1 = LinearRegression()
```

```python
#fit the model to the training dataset
linear_model1.fit(x_train, y_train)

#obtain predictions
predicted_values1 = linear_model1.predict(x_test)
print(predicted_values1)
```

```python
#Actual vs Predicted values
comparison = pd.DataFrame({'Actual':y_test,'Predicted':predicted_values1})
comparison
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values1)
round(mae, 3)
```

```python
#Residual Sum of sqaures(RSS)
import numpy as np

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,predicted_values1))
round(rmse, 3)
```
